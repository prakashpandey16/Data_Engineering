```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
stocks_a = [105, 100, 115, 109, 125, 122, 135]
stocks_b = [123, 128, 120, 134, 134, 148, 113]
stocks_c = [99,91, 100, 112, 109, 119, 124]
```

```python
plt.plot(stocks_a,label = "Company 1")
plt.plot(stocks_b,label = "Company 2")
plt.plot(stocks_c,label = "Company 3")
plt.legend(loc = "upper left")

```
