```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate 1000 random age values following a normal distribution
# Mean (center) = 20, Standard Deviation = 1.5
ages = np.random.normal(20, 1.5, 1000)

# Print generated ages (optional, can be uncommented to see raw values)
# print(ages)

# Create a histogram with custom bins
plt.hist(
    ages,                              # Data to plot (ages)
    bins=[ages.min(), 18, 21, ages.max()],  # Bin edges: Minimum age, 18, 21, Maximum age
)
```
