```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
years = [2012 + x for x in range(2,16,2)]
income = [ 55 + y for y in range(5,40,5)]
income_ticks = list(range(50,101,5))
```

```python
plt.plot(years,income)
plt.title("Income of john in(USD)")
plt.xlabel("Years")
plt.ylabel("Yearly income (In USD)")
plt.yticks(income_ticks,[f"${x}k" for x in income_ticks])
```
