```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array with values from 0 to 99
x = np.arange(100)

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2)

# Plot a sine wave in the top-left subplot (axes[0, 0])
axes[0, 0].plot(x, np.sin(x))
axes[0, 0].set_title("Sine wave")  # Set the title of the sine wave plot

# Plot a cosine wave in the top-right subplot (axes[0, 1])
axes[0, 1].plot(x, np.cos(x))
axes[0, 1].set_title("Cos Wave")  # Set the title of the cosine wave plot

# Plot random values in the bottom-left subplot (axes[1, 0])
axes[1, 0].plot(x, np.random.random(100))  # Plot random numbers between 0 and 1
axes[1, 0].set_title("Random Functions")  # Set the title of the random function plot

# Plot the natural logarithm of x in the bottom-right subplot (axes[1, 1])
axes[1, 1].plot(x, np.log(x + 1))  # Adding 1 to avoid log(0) which is undefined
axes[1, 1].set_title("Log Functions")  # Set the title of the log function plot
axes[1, 1].set_xlabel("Test")  # Label the x-axis for the log plot

# Set the overall title for the figure
fig.suptitle("Four Plots")  # Title for the entire figure (all subplots)

# Adjust layout to avoid overlapping subplots
plt.tight_layout()  # Automatically adjusts spacing for subplots

# Save the figure as a PNG file with higher resolution (dpi = 500) and no transparent background
plt.savefig("fourplots.png", dpi=500, transparent=False)

# Optionally, you can show the plot
plt.show()


```
