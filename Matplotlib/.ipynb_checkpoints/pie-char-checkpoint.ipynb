```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# List of programming languages (Labels for the pie chart)
langs = ["C++", "Python", "Java", "C#", "C"]

# Corresponding votes for each language
votes = [23, 45, 18, 13, 3]

# Create a pie chart
plt.pie(
    votes,                  # Data values (votes for each language)
    labels=langs,           # Labels for each section of the pie chart
    autopct="%.2f%%",       # Display percentage values with two decimal places
    pctdistance=1.35        # Distance of the percentage labels from the center
)

```
